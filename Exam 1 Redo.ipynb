```python
import math
from collections import OrderedDict as odic

def problem2():
    def loaded(path):
        
        #load data
        def getFields(line):
            
            return [field.strip("\n").replace("Iris-","") for field in line.split(",")]        
        
        # Geting header information
        with open(path, "r") as file:
            lines = file.readlines()
            headers = [field for field in getFields(lines[0])]
            values = [dict(zip(headers, getFields(line)))\
                           for line in lines[1:]]
        return headers, values            
        
    def processed(headers, values):
       
        def processField(values, header):
            
            def count(values):
                
                result = 0
                for value in values:
                    result+=1
                return result
            def add(values):
                
                result = 0
                for value in values:
                    result+=value
                return result
            def minimum(values):
                
                result = values[0]
                for value in values:
                    result = value if value<result else result
                return result
            def maximum(values):
               
                result = values[0]
                for value in values:
                    result = value if value>result else result
                return result
            def mean(values):
                
                return add(values)/count(values)
            def variance(values):
               
                cnt = count(values)
                mn = mean(values)
                return add(map(lambda v : (v-mn)**2/cnt, values))
            def std(values):
                
                return math.sqrt(variance(values))            
            
            return odic([("cnt",count(values))\
                    ,("sum",add(values))\
                    ,("min",minimum(values))\
                    ,("max",maximum(values))\
                    ,("mean",mean(values))\
                    ,("var",variance(values))\
                    ,("std",std(values))\
                    ,("header",header)])

        def processSp(sp,headers,values):
            
            return [ processField([float(value[h]) for value in values if value["Species"]==sp],h)\
                     for h in headers if h != "Species" ] 
       
        species = set([x["Species"] for x in values])
        
       
        return [{sp:processSp(sp, headers, values)} for sp in species] 

    def output(x):
       
        def show(species, stats):
            

            def formatStatLine(statName, statResult):
                
                return '{:9} {:6.2f} '.format("{}:".format(statName), statResult)

            def getHeaders():
                
                return [x['header'] for x in stats]

            def formatHeaders():
                
                return "{:11}{}".format(" ","".join(["|{:17}".format(h) for h in getHeaders()]))

            def formatSpecies():
                
                return "{}{}".format(
                            "{:11}".format(species),\
                            # after the species name, show a deviding line (i.e. '======')
                            "{}".format("".join(["|{}".format("="*17) for h in getHeaders()])))

            def formatStats():
                
                # get names
                statNames = []
                [statNames.append(key) for stat in stats for key in stat.keys() \
                     if key != 'header' and key not in statNames]
                return "\n".join(\
                          ["{:11}|{}".format(" ",\
                            "|".join(\
                            [formatStatLine(key,stat[key]) for stat in stats]\
                            )\
                           )\
                          for key in statNames]\
                        )

            # Print headers
            print(formatHeaders())

            # Print species 
            print(formatSpecies())

            # Print statistics
            print(formatStats())

            # Print a line
            print()        
        
        # Showing statistics
        [show(sp, speciesDicionary[sp])\
             for speciesDicionary in x\
             for sp in speciesDicionary.keys()]

    # Printing 
    output(processed(*loaded("iris.txt")))
problem2()
```

```python
from collections import Counter 
import math
import string

def problem3():
    
    def removeNonAlpha(word):
        # Remove nonalpha words
        
        return "".join([c for c in word if c.isalpha()])
    
    def loaded(path):
        
        # Loading 
        
        with open(path) as file:
            # Reading words
            
            words = [removeNonAlpha(word) for line in file for word in line.split(" ")] 
            # Return words 
            
            return [word for word in words if word != ""]
    
    def processed(words):
        
        
        # 10 most frequest words
        freqWords = [(word,freq) for (word,freq) in Counter(words).most_common(10)]
        
        # deduplication
        dedupWords = set(words)
        
        # 10 longest words
        longestWords = [(word,len(word)) for word in\
                        sorted(dedupWords,key=lambda x : len(x), reverse=True)]\
                        [:10]
            
        # Mean        
        lens = [len(word) for word in dedupWords]
        mean = sum(lens)/len(dedupWords)
        
        # Standard Deviation
        std = math.sqrt(sum(map(lambda t : (t-mean)**2/len(lens), lens)))
        
        # Return
        return freqWords, longestWords, mean, std 
    
    def output(frq,lng,mn,std):
       
        
        wordLen = max([length for word, length in lng ])+2

        print("Most Common Words:")
        print("")
        print("")
        print("{w:<{l}}|{c}"\
                         .format(w="Word"\
                                 , l=wordLen\
                                 , c="freq"))
        print("..................................")
        print("\n".join(["{w:<{l}}|{c}"\
                         .format(w="{}:".format(x)\
                                 , l=wordLen\
                                 , c=y)\
                         for x,y in frq]))
        
        print("..................................")
        print()
        print()
        
         # Print
        print("Longest Words:")
        print()
        print("{w:<{l}}|{c}"\
                         .format(w="Word"\
                                 , l=wordLen\
                                 , c="Length"))
        print("..................................")       
        print("\n".join(["{w:<{l}}|{c}"\
                         .format(w="{}:".format(x), l=wordLen, c=y)\
                         for x,y in lng]))
        
        # Print
        print("..................................")
        print("Mean Word Length:     {:.2f}".format(mn))
        
        # Print
        print("..................................")
        print("Word Length Std. Dev: {:.2f}".format(std))
        
    output(*processed(loaded("bible.txt")))
    
problem3()
```
